```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform
import os
from pathlib import Path

# 한글 폰트 설정 함수 (개선된 버전)
def setup_korean_font():
    """
    운영체제별로 한글 폰트를 설정합니다.
    """
    system = platform.system()
    
    # 사용 가능한 한글 폰트 찾기
    font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
    korean_fonts = []
    
    for font_path in font_list:
        try:
            font_name = fm.FontProperties(fname=font_path).get_name()
            # 한글 폰트 이름 패턴 확인
            if any(keyword in font_name.upper() for keyword in 
                   ['GOTHIC', 'MALGUN', 'NANUM', 'DOTUM', 'GULIM', 'BATANG']):
                korean_fonts.append(font_name)
        except:
            continue
    
    # 운영체제별 우선순위 폰트 설정
    if system == 'Windows':
        preferred_fonts = ['Malgun Gothic', 'NanumGothic', 'Dotum']
    elif system == 'Darwin':  # macOS
        preferred_fonts = ['AppleGothic', 'NanumGothic']
    else:  # Linux
        preferred_fonts = ['NanumGothic', 'DejaVu Sans']
    
    # 사용 가능한 폰트 중에서 우선순위대로 설정
    font_set = False
    for font in preferred_fonts:
        if font in korean_fonts:
            plt.rcParams['font.family'] = font
            font_set = True
            print(f"한글 폰트 설정 완료: {font}")
            break
    
    if not font_set:
        # 사용 가능한 한글 폰트가 있다면 첫 번째 것 사용
        if korean_fonts:
            plt.rcParams['font.family'] = korean_fonts[0]
            print(f"한글 폰트 설정 완료: {korean_fonts[0]}")
        else:
            print("한글 폰트를 찾을 수 없습니다. 기본 폰트를 사용합니다.")
    
    # 마이너스 기호 깨짐 방지
    plt.rcParams['axes.unicode_minus'] = False

# 폰트 설정 실행
setup_korean_font()

class MusicScoreProcessor:
    def __init__(self, image_path):
        """
        악보 이미지 처리를 위한 클래스
        """
        self.image_path = image_path
        self.original_image = None
        self.binary_image = None
        self.note_contours = []
        self.note_images = []
        
    def validate_image_path(self):
        """
        이미지 경로를 검증하고 다양한 형식을 지원합니다.
        """
        # 경로 정규화
        path = Path(self.image_path)
        
        print(f"입력된 경로: {self.image_path}")
        print(f"정규화된 경로: {path}")
        print(f"절대 경로: {path.absolute()}")
        print(f"경로 존재 여부: {path.exists()}")
        
        if path.exists():
            print(f"파일 크기: {path.stat().st_size} bytes")
            return str(path)
        
        # 확장자가 없는 경우 다양한 확장자로 시도
        if not path.suffix:
            extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif']
            for ext in extensions:
                test_path = path.with_suffix(ext)
                print(f"확장자 시도: {test_path}")
                if test_path.exists():
                    print(f"찾은 파일: {test_path}")
                    return str(test_path)
        
        # 같은 디렉토리에서 파일명만으로 검색
        if path.name:
            current_dir = Path.cwd()
            for file in current_dir.glob(f"*{path.stem}*"):
                if file.suffix.lower() in ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif']:
                    print(f"현재 디렉토리에서 찾은 파일: {file}")
                    return str(file)
        
        return None
        
    def load_image(self):
        """
        악보 이미지를 불러옵니다.
        """
        # 경로 검증
        validated_path = self.validate_image_path()
        
        if validated_path is None:
            print("이미지 파일을 찾을 수 없어 샘플 악보를 생성합니다.")
            self.create_sample_score()
            return True
        
        # 다양한 방법으로 이미지 로드 시도
        try:
            # 방법 1: 기본 imread
            self.original_image = cv2.imread(validated_path, cv2.IMREAD_GRAYSCALE)
            
            if self.original_image is not None:
                print(f"이미지 로드 성공: {validated_path}")
                print(f"이미지 크기: {self.original_image.shape}")
                return True
            
            # 방법 2: 유니코드 경로 문제 해결
            # OpenCV는 한글 경로를 제대로 처리하지 못할 수 있음
            with open(validated_path, 'rb') as f:
                image_data = f.read()
            
            # numpy 배열로 변환 후 OpenCV로 디코딩
            nparr = np.frombuffer(image_data, np.uint8)
            self.original_image = cv2.imdecode(nparr, cv2.IMREAD_GRAYSCALE)
            
            if self.original_image is not None:
                print(f"유니코드 경로 처리로 이미지 로드 성공: {validated_path}")
                print(f"이미지 크기: {self.original_image.shape}")
                return True
            
        except Exception as e:
            print(f"이미지 로드 중 오류 발생: {e}")
        
        print("이미지 로드에 실패하여 샘플 악보를 생성합니다.")
        self.create_sample_score()
        return True
    
    def create_sample_score(self):
        """
        테스트용 샘플 악보 이미지를 생성합니다.
        """
        # 흰색 배경 생성
        self.original_image = np.ones((400, 800), dtype=np.uint8) * 255
        
        # 오선 그리기
        for i in range(5):
            y = 100 + i * 20
            cv2.line(self.original_image, (50, y), (750, y), 0, 2)
        
        # 샘플 음표들 그리기 (원형으로)
        note_positions = [
            (100, 90), (150, 110), (200, 130), (250, 120), (300, 100),
            (350, 140), (400, 110), (450, 90), (500, 130), (550, 120),
            (600, 100), (650, 110), (700, 90)
        ]
        
        for x, y in note_positions:
            cv2.circle(self.original_image, (x, y), 8, 0, -1)  # 채워진 원
            cv2.line(self.original_image, (x+8, y), (x+8, y-40), 0, 2)  # 음표 줄기
    
    def preprocess_image(self):
        """
        이미지 전처리 - 이진화 및 노이즈 제거
        """
        # 가우시안 블러로 노이즈 제거
        blurred = cv2.GaussianBlur(self.original_image, (3, 3), 0)
        
        # 적응형 임계값을 사용한 이진화
        self.binary_image = cv2.adaptiveThreshold(
            blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
            cv2.THRESH_BINARY_INV, 11, 2
        )
        
        # 모폴로지 연산으로 노이즈 제거
        kernel = np.ones((2, 2), np.uint8)
        self.binary_image = cv2.morphologyEx(self.binary_image, cv2.MORPH_CLOSE, kernel)
        
    def detect_notes(self):
        """
        음표 위치를 감지합니다.
        """
        # 컨투어 검출
        contours, _ = cv2.findContours(
            self.binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
        )
        
        print(f"전체 컨투어 수: {len(contours)}")
        
        # 음표로 추정되는 컨투어 필터링 (더 관대한 조건)
        min_area = 20  # 최소 면적 줄임
        max_area = 5000  # 최대 면적 늘림
        min_aspect_ratio = 0.1  # 더 관대한 종횡비
        max_aspect_ratio = 10.0
        
        for i, cnt in enumerate(contours):
            area = cv2.contourArea(cnt)
            if area < min_area or area > max_area:
                continue
                
            x, y, w, h = cv2.boundingRect(cnt)
            aspect_ratio = w / h if h > 0 else 0
            
            print(f"컨투어 {i}: 면적={area:.1f}, 크기=({w}x{h}), 종횡비={aspect_ratio:.2f}")
            
            if min_aspect_ratio <= aspect_ratio <= max_aspect_ratio:
                self.note_contours.append((x, y, w, h, area))
        
        # 위치에 따라 정렬 (위→아래, 왼쪽→오른쪽)
        self.note_contours.sort(key=lambda note: (note[1] // 30, note[0]))
        
        print(f"감지된 음표 수: {len(self.note_contours)}")
        
    def extract_note_images(self, target_size=(299, 299)):
        """
        감지된 음표들을 개별 이미지로 추출하고 크기를 조정합니다.
        """
        self.note_images = []
        
        for i, (x, y, w, h, area) in enumerate(self.note_contours):
            # 여백을 포함하여 음표 영역 확장
            margin = 10
            x1 = max(0, x - margin)
            y1 = max(0, y - margin)
            x2 = min(self.original_image.shape[1], x + w + margin)
            y2 = min(self.original_image.shape[0], y + h + margin)
            
            # 음표 이미지 추출
            note_img = self.original_image[y1:y2, x1:x2]
            
            # 299x299 크기로 리사이즈
            resized_note = cv2.resize(note_img, target_size, interpolation=cv2.INTER_AREA)
            
            # 정규화 (0-1 범위로)
            normalized_note = resized_note.astype(np.float32) / 255.0
            
            self.note_images.append({
                'image': normalized_note,
                'original_bbox': (x, y, w, h),
                'position': i,
                'area': area
            })
    
    def visualize_detection(self):
        """
        음표 감지 결과를 시각화합니다.
        """
        # 원본 이미지에 감지된 음표 표시
        image_color = cv2.cvtColor(self.original_image, cv2.COLOR_GRAY2BGR)
        
        for i, (x, y, w, h, area) in enumerate(self.note_contours):
            # 바운딩 박스 그리기
            cv2.rectangle(image_color, (x, y), (x+w, y+h), (0, 255, 0), 2)
            # 순서 번호 표시
            cv2.putText(image_color, str(i+1), (x, y-5), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
        
        plt.figure(figsize=(15, 8))
        
        # 원본 이미지와 감지 결과
        plt.subplot(2, 2, 1)
        plt.imshow(self.original_image, cmap='gray')
        plt.title('원본 악보 이미지')
        plt.axis('off')
        
        plt.subplot(2, 2, 2)
        plt.imshow(self.binary_image, cmap='gray')
        plt.title('이진화된 이미지')
        plt.axis('off')
        
        plt.subplot(2, 2, 3)
        plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
        plt.title(f'감지된 음표 ({len(self.note_contours)}개)')
        plt.axis('off')
        
        # 추출된 음표 샘플 표시
        plt.subplot(2, 2, 4)
        if self.note_images:
            sample_notes = self.note_images[:6]  # 처음 6개만 표시
            combined_img = np.hstack([note['image'] for note in sample_notes])
            plt.imshow(combined_img, cmap='gray')
            plt.title('추출된 음표 샘플 (299x299)')
        else:
            plt.text(0.5, 0.5, '감지된 음표가 없습니다', ha='center', va='center')
            plt.title('음표 없음')
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
    
    def get_model_input_data(self):
        """
        모델 입력용 데이터를 반환합니다.
        """
        if not self.note_images:
            return None
            
        # 모든 음표 이미지를 배열로 변환
        input_array = np.array([note['image'] for note in self.note_images])
        
        # 채널 차원 추가 (그레이스케일이므로 1채널)
        if len(input_array.shape) == 3:
            input_array = np.expand_dims(input_array, axis=-1)
        
        return {
            'images': input_array,
            'positions': [note['position'] for note in self.note_images],
            'bboxes': [note['original_bbox'] for note in self.note_images],
            'shape': input_array.shape
        }
    
    def process_score(self):
        """
        전체 악보 처리 파이프라인을 실행합니다.
        """
        print("=== 악보 이미지 처리 시작 ===")
        
        # 1. 이미지 로드
        if not self.load_image():
            print("이미지 로드 실패")
            return None
        print("✓ 이미지 로드 완료")
        
        # 2. 전처리
        self.preprocess_image()
        print("✓ 이미지 전처리 완료")
        
        # 3. 음표 감지
        self.detect_notes()
        print("✓ 음표 감지 완료")
        
        # 4. 음표 이미지 추출
        self.extract_note_images()
        print("✓ 음표 이미지 추출 완료")
        
        # 5. 시각화
        self.visualize_detection()
        
        # 6. 모델 입력 데이터 준비
        model_data = self.get_model_input_data()
        
        print("=== 처리 완료 ===")
        print(f"추출된 음표 수: {len(self.note_images)}")
        print(f"모델 입력 데이터 형태: {model_data['shape'] if model_data else 'None'}")
        
        return model_data

# 사용 예시
if __name__ == "__main__":
    # 악보 처리기 생성
    image_path = r"C:\Users\kimsi\Desktop\명지대\딥러닝 과목 프로젝트\korea.jpg"
    processor = MusicScoreProcessor(image_path)
    
    # 전체 파이프라인 실행
    model_input_data = processor.process_score()
    
    if model_input_data:
        print("\n=== 모델 입력 준비 완료 ===")
        print(f"입력 이미지 배열 크기: {model_input_data['images'].shape}")
        print(f"음표 위치 정보: {len(model_input_data['positions'])}개")
        
        # 실제 모델 예측 시뮬레이션
        print("\n=== 모델 예측 시뮬레이션 ===")
        for i, pos in enumerate(model_input_data['positions']):
            print(f"음표 {i+1}: 위치 순서 {pos+1}")
```

```python
import cv2
import numpy as np
from pathlib import Path

class MusicScoreProcessor:
    def __init__(self, image_path):
        """
        악보 이미지 처리를 위한 클래스
        """
        self.image_path = image_path
        self.original_image = None
        self.binary_image = None
        self.note_contours = []
        self.note_images = []
        
    def validate_image_path(self):
        """
        이미지 경로를 검증하고 다양한 형식을 지원합니다.
        """
        path = Path(self.image_path)
        
        if path.exists():
            return str(path)
        
        # 확장자가 없는 경우 다양한 확장자로 시도
        if not path.suffix:
            extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif']
            for ext in extensions:
                test_path = path.with_suffix(ext)
                if test_path.exists():
                    return str(test_path)
        
        return None
        
    def load_image(self):
        """
        악보 이미지를 불러옵니다.
        """
        validated_path = self.validate_image_path()
        
        if validated_path is None:
            return False
        
        try:
            # 방법 1: 기본 imread
            self.original_image = cv2.imread(validated_path, cv2.IMREAD_GRAYSCALE)
            
            if self.original_image is not None:
                return True
            
            # 방법 2: 유니코드 경로 문제 해결
            with open(validated_path, 'rb') as f:
                image_data = f.read()
            
            nparr = np.frombuffer(image_data, np.uint8)
            self.original_image = cv2.imdecode(nparr, cv2.IMREAD_GRAYSCALE)
            
            if self.original_image is not None:
                return True
            
        except Exception as e:
            print(f"이미지 로드 중 오류 발생: {e}")
        
        return False
    
    def preprocess_image(self):
        """
        이미지 전처리 - 이진화 및 노이즈 제거
        """
        blurred = cv2.GaussianBlur(self.original_image, (3, 3), 0)
        
        self.binary_image = cv2.adaptiveThreshold(
            blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
            cv2.THRESH_BINARY_INV, 11, 2
        )
        
        kernel = np.ones((2, 2), np.uint8)
        self.binary_image = cv2.morphologyEx(self.binary_image, cv2.MORPH_CLOSE, kernel)
        
    def detect_notes(self):
        """
        음표 위치를 감지합니다.
        """
        contours, _ = cv2.findContours(
            self.binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
        )
        
        min_area = 20
        max_area = 5000
        min_aspect_ratio = 0.1
        max_aspect_ratio = 10.0
        
        for cnt in contours:
            area = cv2.contourArea(cnt)
            if area < min_area or area > max_area:
                continue
                
            x, y, w, h = cv2.boundingRect(cnt)
            aspect_ratio = w / h if h > 0 else 0
            
            if min_aspect_ratio <= aspect_ratio <= max_aspect_ratio:
                self.note_contours.append((x, y, w, h, area))
        
        # 위치에 따라 정렬 (위→아래, 왼쪽→오른쪽)
        self.note_contours.sort(key=lambda note: (note[1] // 30, note[0]))
        
    def extract_resized_notes(self, target_size=(299, 299)):
        """
        299x299 크기로 리사이즈된 음표 이미지들만 반환합니다.
        """
        resized_notes = []
        
        for i, (x, y, w, h, area) in enumerate(self.note_contours):
            # 여백을 포함하여 음표 영역 확장
            margin = 10
            x1 = max(0, x - margin)
            y1 = max(0, y - margin)
            x2 = min(self.original_image.shape[1], x + w + margin)
            y2 = min(self.original_image.shape[0], y + h + margin)
            
            # 음표 이미지 추출
            note_img = self.original_image[y1:y2, x1:x2]
            
            # 299x299 크기로 리사이즈
            resized_note = cv2.resize(note_img, target_size, interpolation=cv2.INTER_AREA)
            
            # 정규화 (0-1 범위로)
            normalized_note = resized_note.astype(np.float32) / 255.0
            
            resized_notes.append(normalized_note)
        
        return resized_notes
    
    def get_resized_note_images(self):
        """
        전체 파이프라인을 실행하고 299x299 크기의 음표 이미지들만 반환합니다.
        """
        print("음표 이미지 추출 시작...")
        
        # 1. 이미지 로드
        if not self.load_image():
            print("이미지 로드 실패")
            return []
        
        # 2. 전처리
        self.preprocess_image()
        
        # 3. 음표 감지
        self.detect_notes()
        
        # 4. 299x299 크기로 리사이즈된 음표들만 추출
        resized_notes = self.extract_resized_notes()
        
        print(f"추출된 299x299 음표 이미지 수: {len(resized_notes)}")
        
        return resized_notes

# 사용 예시
if __name__ == "__main__":
    # 악보 처리기 생성
    image_path = r"C:\Users\kimsi\Desktop\명지대\딥러닝 과목 프로젝트\korea.jpg"
    processor = MusicScoreProcessor(image_path)
    
    # 299x299 크기의 음표 이미지들만 추출
    resized_note_images = processor.get_resized_note_images()
    
    if resized_note_images:
        print(f"\n=== 결과 ===")
        print(f"추출된 이미지 수: {len(resized_note_images)}")
        print(f"각 이미지 크기: {resized_note_images[0].shape}")
        print(f"데이터 타입: {resized_note_images[0].dtype}")
        print(f"값 범위: {resized_note_images[0].min():.3f} ~ {resized_note_images[0].max():.3f}")
        
        # numpy 배열로 변환 (모델 입력용)
        note_array = np.array(resized_note_images)
        print(f"전체 배열 크기: {note_array.shape}")
        
        # 모델에 바로 입력할 수 있는 형태로 변환
        # 채널 차원 추가 (그레이스케일이므로 1채널)
        model_input = np.expand_dims(note_array, axis=-1)
        print(f"모델 입력 형태: {model_input.shape}")
    else:
        print("음표를 찾을 수 없습니다.")
```

```python
import cv2
import numpy as np
import os
from pathlib import Path

class MusicScoreProcessor:
    def __init__(self, image_path, output_dir=None):
        """
        악보 이미지 처리를 위한 클래스
        """
        self.image_path = image_path
        
        # 절대 경로로 출력 디렉토리 설정
        if output_dir is None:
            # 이미지 파일과 같은 디렉토리에 저장
            image_parent_dir = Path(image_path).parent
            self.output_dir = image_parent_dir / "extracted_notes_output"
        else:
            self.output_dir = Path(output_dir)
        
        # 출력 디렉토리 생성
        self.output_dir.mkdir(exist_ok=True)
        
        print(f"출력 디렉토리: {self.output_dir.absolute()}")
        
        self.original_image = None
        self.binary_image = None
        self.note_contours = []
        self.note_images = []
        
    def validate_image_path(self):
        """
        이미지 경로를 검증하고 다양한 형식을 지원합니다.
        """
        path = Path(self.image_path)
        
        if path.exists():
            return str(path)
        
        # 확장자가 없는 경우 다양한 확장자로 시도
        if not path.suffix:
            extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.tif']
            for ext in extensions:
                test_path = path.with_suffix(ext)
                if test_path.exists():
                    return str(test_path)
        
        return None
        
    def load_image(self):
        """
        악보 이미지를 불러옵니다.
        """
        validated_path = self.validate_image_path()
        
        if validated_path is None:
            return False
        
        try:
            # 방법 1: 기본 imread
            self.original_image = cv2.imread(validated_path, cv2.IMREAD_GRAYSCALE)
            
            if self.original_image is not None:
                return True
            
            # 방법 2: 유니코드 경로 문제 해결
            with open(validated_path, 'rb') as f:
                image_data = f.read()
            
            nparr = np.frombuffer(image_data, np.uint8)
            self.original_image = cv2.imdecode(nparr, cv2.IMREAD_GRAYSCALE)
            
            if self.original_image is not None:
                return True
            
        except Exception as e:
            print(f"이미지 로드 중 오류 발생: {e}")
        
        return False
    
    def preprocess_image(self):
        """
        이미지 전처리 - 이진화 및 노이즈 제거
        """
        blurred = cv2.GaussianBlur(self.original_image, (3, 3), 0)
        
        self.binary_image = cv2.adaptiveThreshold(
            blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
            cv2.THRESH_BINARY_INV, 11, 2
        )
        
        kernel = np.ones((2, 2), np.uint8)
        self.binary_image = cv2.morphologyEx(self.binary_image, cv2.MORPH_CLOSE, kernel)
        
    def detect_notes(self):
        """
        음표 위치를 감지합니다.
        """
        contours, _ = cv2.findContours(
            self.binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
        )
        
        min_area = 20
        max_area = 5000
        min_aspect_ratio = 0.1
        max_aspect_ratio = 10.0
        
        for cnt in contours:
            area = cv2.contourArea(cnt)
            if area < min_area or area > max_area:
                continue
                
            x, y, w, h = cv2.boundingRect(cnt)
            aspect_ratio = w / h if h > 0 else 0
            
            if min_aspect_ratio <= aspect_ratio <= max_aspect_ratio:
                self.note_contours.append((x, y, w, h, area))
        
        # 위치에 따라 정렬 (위→아래, 왼쪽→오른쪽)
        self.note_contours.sort(key=lambda note: (note[1] // 30, note[0]))
        
    def extract_and_save_notes(self, target_size=(299, 299)):
        """
        299x299 크기로 리사이즈된 음표 이미지들을 추출하고 파일로 저장합니다.
        """
        resized_notes = []
        saved_files = []
        
        for i, (x, y, w, h, area) in enumerate(self.note_contours):
            # 여백을 포함하여 음표 영역 확장
            margin = 10
            x1 = max(0, x - margin)
            y1 = max(0, y - margin)
            x2 = min(self.original_image.shape[1], x + w + margin)
            y2 = min(self.original_image.shape[0], y + h + margin)
            
            # 음표 이미지 추출
            note_img = self.original_image[y1:y2, x1:x2]
            
            # 299x299 크기로 리사이즈
            resized_note = cv2.resize(note_img, target_size, interpolation=cv2.INTER_AREA)
            
            # 파일명 생성 (순서대로 번호 매기기)
            filename = f"note_{i+1:03d}.png"
            filepath = self.output_dir / filename
            
            # PNG 파일로 저장 (절대 경로 사용)
            success = cv2.imwrite(str(filepath), resized_note)
            
            if success:
                print(f"저장됨: {filename} -> {filepath}")
                saved_files.append(str(filepath))
            else:
                print(f"저장 실패: {filename}")
            
            # 정규화된 버전도 메모리에 저장
            normalized_note = resized_note.astype(np.float32) / 255.0
            resized_notes.append(normalized_note)
        
        return resized_notes, saved_files
    
    def save_as_numpy_array(self, resized_notes):
        """
        모든 음표 이미지를 하나의 numpy 배열로 저장합니다.
        """
        if not resized_notes:
            return None
            
        # numpy 배열로 변환
        note_array = np.array(resized_notes)
        
        # 채널 차원 추가 (그레이스케일이므로 1채널)
        model_input = np.expand_dims(note_array, axis=-1)
        
        # numpy 파일로 저장 (절대 경로 사용)
        numpy_filepath = self.output_dir / "all_notes.npy"
        np.save(str(numpy_filepath), model_input)
        
        print(f"전체 배열 저장됨: {numpy_filepath}")
        print(f"배열 크기: {model_input.shape}")
        
        return str(numpy_filepath)
    
    def check_saved_files(self):
        """
        저장된 파일들을 확인합니다.
        """
        if self.output_dir.exists():
            files = list(self.output_dir.glob("*"))
            print(f"\n=== 저장된 파일 확인 ===")
            print(f"출력 디렉토리: {self.output_dir.absolute()}")
            print(f"저장된 파일 수: {len(files)}")
            
            # PNG 파일들
            png_files = list(self.output_dir.glob("*.png"))
            print(f"PNG 파일 수: {len(png_files)}")
            
            # NPY 파일들
            npy_files = list(self.output_dir.glob("*.npy"))
            print(f"NPY 파일 수: {len(npy_files)}")
            
            # 처음 5개 파일 출력
            for i, file in enumerate(files[:5]):
                print(f"  {file.name}")
            if len(files) > 5:
                print(f"  ... 총 {len(files)}개")
        else:
            print("출력 디렉토리가 존재하지 않습니다.")
    
    def process_and_save_notes(self):
        """
        전체 파이프라인을 실행하고 결과를 파일로 저장합니다.
        """
        print("=== 음표 이미지 추출 및 저장 시작 ===")
        
        # 1. 이미지 로드
        if not self.load_image():
            print("이미지 로드 실패")
            return None, None
        print("✓ 이미지 로드 완료")
        
        # 2. 전처리
        self.preprocess_image()
        print("✓ 이미지 전처리 완료")
        
        # 3. 음표 감지
        self.detect_notes()
        print(f"✓ 음표 감지 완료: {len(self.note_contours)}개")
        
        # 4. 299x299 크기로 리사이즈하고 파일로 저장
        resized_notes, saved_files = self.extract_and_save_notes()
        print(f"✓ 개별 이미지 저장 완료: {len(saved_files)}개")
        
        # 5. 전체 배열을 numpy 파일로 저장
        numpy_filepath = self.save_as_numpy_array(resized_notes)
        print("✓ numpy 배열 저장 완료")
        
        # 6. 저장된 파일 확인
        self.check_saved_files()
        
        print("=== 처리 완료 ===")
        
        return saved_files, numpy_filepath

# 사용 예시
if __name__ == "__main__":
    # 악보 처리기 생성 (절대 경로로 출력 디렉토리 지정)
    image_path = r"C:\Users\kimsi\Desktop\명지대\딥러닝 과목 프로젝트\korea.jpg"
    output_directory = r"C:\Users\kimsi\Desktop\명지대\딥러닝 과목 프로젝트\extracted_notes_output"
    
    processor = MusicScoreProcessor(image_path, output_directory)
    
    # 음표 추출 및 저장
    saved_files, numpy_file = processor.process_and_save_notes()
```
