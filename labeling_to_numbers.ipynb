```python
def key_number_to_midi_note(key_number):
    """
    0~87 범위의 피아노 건반 번호를 MIDI 음표 문자열로 변환합니다.
    0은 A0, 87은 C8에 해당합니다.
    """
    note_map = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
    midi_number = key_number + 21
    octave = (midi_number // 12) - 1
    note_index = midi_number % 12
    note_name = note_map[note_index]
    return f"{note_name}{octave}"

# 트랜스포머 모델에서 출력된 숫자 데이터 여기다가 입력
model_output = [62, 62, 62, 54, 36, 36, 70, 70, 70, 51, 70, 29, 55, 29, 62, 70, 38, 62, 20, 70, 36, 36, 87, 62, 62, 62, 81, 62, 72, 8, 87, 70, 62, 54, 70, 70, 36, 62, 70, 62, 62, 70, 62, 36, 70, 76, 32, 23, 36, 73, 57, 62, 38, 70, 70, 62, 40, 14, 70, 70, 70, 60, 70, 70, 62, 36, 57, 70, 87, 20, 62, 62]

# 숫자를 MIDI 음표로 변환
midi_notes = [key_number_to_midi_note(num) for num in model_output]

# 라벨링 결과를 체계적으로 정리
import pandas as pd

labeling_df = pd.DataFrame({
    '순서': range(1, len(model_output) + 1),
    '건반_번호': model_output,
    'MIDI_음표': midi_notes,
    'MIDI_번호': [num + 21 for num in model_output],
    '옥타브': [int(note[-1]) for note in midi_notes],
    '음이름': [note[:-1] for note in midi_notes]
})

print("=== 트랜스포머 모델 출력 라벨링 결과 ===")
print(f"총 음표 수: {len(midi_notes)}")
print(f"사용된 건반 범위: {min(model_output)} ~ {max(model_output)}")
print(f"사용된 옥타브 범위: {min([int(note[-1]) for note in midi_notes])} ~ {max([int(note[-1]) for note in midi_notes])}")

# 전체 라벨링 결과 출력
print("\n=== 전체 라벨링 데이터 ===")
print(labeling_df.to_string(index=False))

# 음표별 빈도 분석
print("\n=== 음표 빈도 분석 ===")
note_frequency = pd.Series(midi_notes).value_counts().head(10)
print("가장 많이 사용된 음표 (상위 10개):")
for note, count in note_frequency.items():
    print(f"{note}: {count}회")

# MIDI 파일 생성을 위한 딕셔너리
note_sequence = {
    'notes': midi_notes,
    'key_numbers': model_output,
    'midi_numbers': [num + 21 for num in model_output]
}

print(f"\n=== MIDI 변환 완료 ===")
print("변환된 음표 시퀀스:")
print(" -> ".join(midi_notes))
```

```python
import sys
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from pathlib import Path
import platform

# 한글 폰트 설정
def setup_korean_font():
    system = platform.system()
    if system == 'Windows':
        plt.rcParams['font.family'] = 'Malgun Gothic'
    elif system == 'Darwin':
        plt.rcParams['font.family'] = 'AppleGothic'
    else:
        plt.rcParams['font.family'] = 'DejaVu Sans'
    plt.rcParams['axes.unicode_minus'] = False

setup_korean_font()

class RelaxedScoreProcessor:
    def __init__(self, image_path, output_dir='relaxed_output'):
        """
        완화된 필터링을 적용한 악보 처리기
        """
        self.image_path = image_path
        self.output_dir = Path(output_dir)
        self.output_dir.mkdir(exist_ok=True)
        
        self.original_image = None
        self.processed_images = {}
        self.staves = []
        self.objects = []
        self.filtered_objects = []
        self.note_images = []
        
        print(f"출력 디렉토리: {self.output_dir.absolute()}")
    
    def load_image(self):
        """이미지 로드"""
        try:
            self.original_image = cv2.imread(self.image_path)
            if self.original_image is None:
                with open(self.image_path, 'rb') as f:
                    image_data = f.read()
                nparr = np.frombuffer(image_data, np.uint8)
                self.original_image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
            return True
        except Exception as e:
            print(f"이미지 로드 실패: {e}")
            return False
    
    def simple_preprocessing(self):
        """단순한 전처리"""
        gray = cv2.cvtColor(self.original_image, cv2.COLOR_BGR2GRAY)
        
        # 적응형 임계값
        binary = cv2.adaptiveThreshold(
            gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
            cv2.THRESH_BINARY_INV, 11, 2
        )
        
        # 작은 노이즈만 제거
        kernel = np.ones((2, 2), np.uint8)
        cleaned = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)
        
        self.processed_images['preprocessed'] = cleaned
        return cleaned
    
    def relaxed_object_detection(self, image):
        """완화된 객체 검출"""
        contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        objects = []
        for i, contour in enumerate(contours):
            x, y, w, h = cv2.boundingRect(contour)
            area = cv2.contourArea(contour)
            
            # 매우 관대한 필터링 - 거의 모든 객체 허용
            if area > 10 and w > 2 and h > 2:  # 최소 조건만
                objects.append([i, [x, y, w, h, area]])
        
        print(f"검출된 전체 객체 수: {len(objects)}")
        return objects
    
    def very_relaxed_filtering(self, objects):
        """매우 완화된 필터링"""
        print(f"필터링 전 객체 수: {len(objects)}")
        
        filtered = []
        
        for obj in objects:
            obj_id, bbox = obj[0], obj[1]
            x, y, w, h = bbox[:4]
            area = bbox[4] if len(bbox) > 4 else w * h
            
            # 1. 매우 큰 객체만 제외 (전체 이미지의 1/4 이상)
            image_area = self.original_image.shape[0] * self.original_image.shape[1]
            if area > image_area * 0.25:
                print(f"너무 큰 객체 제외: 면적={area}")
                continue
            
            # 2. 매우 작은 점만 제외
            if area < 5 or w < 2 or h < 2:
                continue
            
            # 3. 극단적인 종횡비만 제외
            aspect_ratio = w / h if h > 0 else 0
            if aspect_ratio > 20.0 or aspect_ratio < 0.05:
                continue
            
            filtered.append(obj)
        
        print(f"완화된 필터링 후 객체 수: {len(filtered)}")
        
        # 위치 기반 정렬
        filtered.sort(key=lambda obj: (obj[1][1] // 30, obj[1][0]))
        
        self.filtered_objects = filtered
        return filtered
    
    def extract_all_objects(self, objects, target_size=299):
        """모든 객체를 추출하여 저장"""
        note_images = []
        saved_files = []
        
        for i, obj in enumerate(objects):
            try:
                x, y, w, h = obj[1][:4]
                
                # 여백 추가
                margin = 10
                x1 = max(0, x - margin)
                y1 = max(0, y - margin)
                x2 = min(self.original_image.shape[1], x + w + margin)
                y2 = min(self.original_image.shape[0], y + h + margin)
                
                # 크롭
                crop = self.original_image[y1:y2, x1:x2]
                if crop.size == 0:
                    continue
                
                # 그레이스케일 처리
                if len(crop.shape) == 3:
                    crop = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
                
                # 리사이즈
                resized = self.resize_with_padding(crop, target_size)
                
                # 파일 저장
                filename = f"object_{i+1:03d}.png"
                filepath = self.output_dir / filename
                
                success = cv2.imwrite(str(filepath), resized)
                if success:
                    print(f"저장됨: {filename}")
                    saved_files.append(str(filepath))
                    note_images.append(resized)
                
            except Exception as e:
                print(f"객체 {i} 처리 중 오류: {e}")
                continue
        
        self.note_images = note_images
        return note_images, saved_files
    
    def resize_with_padding(self, image, target_size):
        """비율 유지하면서 패딩 추가"""
        h, w = image.shape[:2]
        
        scale = min(target_size / w, target_size / h)
        new_w = int(w * scale)
        new_h = int(h * scale)
        
        resized = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_AREA)
        
        top = (target_size - new_h) // 2
        bottom = target_size - new_h - top
        left = (target_size - new_w) // 2
        right = target_size - new_w - left
        
        padded = cv2.copyMakeBorder(
            resized, top, bottom, left, right, 
            cv2.BORDER_CONSTANT, value=255
        )
        
        return padded
    
    def visualize_results(self):
        """수정된 결과 시각화"""
        # 첫 번째 figure: 전체 처리 과정
        plt.figure(figsize=(15, 5))
        
        # 원본 이미지
        plt.subplot(1, 3, 1)
        plt.imshow(cv2.cvtColor(self.original_image, cv2.COLOR_BGR2RGB))
        plt.title('원본 이미지')
        plt.axis('off')
        
        # 전처리된 이미지
        if 'preprocessed' in self.processed_images:
            plt.subplot(1, 3, 2)
            plt.imshow(self.processed_images['preprocessed'], cmap='gray')
            plt.title('전처리된 이미지')
            plt.axis('off')
        
        # 감지된 객체들
        result_img = self.original_image.copy()
        for i, obj in enumerate(self.filtered_objects):
            x, y, w, h = obj[1][:4]
            cv2.rectangle(result_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
            cv2.putText(result_img, str(i+1), (x, y-5), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
        
        plt.subplot(1, 3, 3)
        plt.imshow(cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB))
        plt.title(f'감지된 객체 ({len(self.filtered_objects)}개)')
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # 두 번째 figure: 추출된 객체들 (별도로 표시)
        if self.note_images:
            # 한 번에 표시할 수 있는 최대 개수 계산
            max_display = min(12, len(self.note_images))  # 최대 12개
            
            if max_display > 0:
                # 그리드 크기 계산
                cols = 4
                rows = (max_display + cols - 1) // cols  # 올림 계산
                
                plt.figure(figsize=(16, 4 * rows))
                
                for i in range(max_display):
                    plt.subplot(rows, cols, i + 1)
                    plt.imshow(self.note_images[i], cmap='gray')
                    plt.title(f'객체 {i+1}')
                    plt.axis('off')
                
                plt.tight_layout()
                plt.show()
                
                # 나머지 객체들이 있다면 알림
                if len(self.note_images) > max_display:
                    print(f"처음 {max_display}개만 표시됨. 총 {len(self.note_images)}개 객체가 저장되었습니다.")
    
    def process_with_relaxed_filtering(self):
        """완화된 필터링으로 처리"""
        print("=== 완화된 악보 처리 시작 ===")
        
        if not self.load_image():
            return None
        print("✓ 이미지 로드 완료")
        
        # 단순 전처리
        processed = self.simple_preprocessing()
        print("✓ 전처리 완료")
        
        # 완화된 객체 검출
        objects = self.relaxed_object_detection(processed)
        print("✓ 객체 검출 완료")
        
        # 매우 완화된 필터링
        filtered_objects = self.very_relaxed_filtering(objects)
        print("✓ 완화된 필터링 완료")
        
        # 모든 객체 추출
        note_images, saved_files = self.extract_all_objects(filtered_objects)
        print("✓ 객체 추출 완료")
        
        # 시각화
        self.visualize_results()
        
        print("=== 처리 완료 ===")
        print(f"감지된 객체 수: {len(objects)}")
        print(f"필터링된 객체 수: {len(filtered_objects)}")
        print(f"저장된 이미지 수: {len(saved_files)}")
        
        # numpy 배열 저장
        if note_images:
            note_array = np.array(note_images)
            normalized_array = note_array.astype(np.float32) / 255.0
            model_input = np.expand_dims(normalized_array, axis=-1)
            
            numpy_path = self.output_dir / "relaxed_objects_array.npy"
            np.save(str(numpy_path), model_input)
            print(f"numpy 배열 저장: {numpy_path}")
            print(f"모델 입력 형태: {model_input.shape}")
        
        return {
            'note_images': note_images,
            'saved_files': saved_files,
            'filtered_objects': filtered_objects
        }

# 사용
if __name__ == "__main__":
    image_path = r'C:\Users\kimsi\Desktop\명지대\딥러닝 과목 프로젝트\korea.jpg'
    processor = RelaxedScoreProcessor(image_path, 'relaxed_output')
    
    results = processor.process_with_relaxed_filtering()
    
    if results and results['note_images']:
        print(f"\n=== 성공! ===")
        print(f"추출된 객체 수: {len(results['note_images'])}")
        print("이제 실제로 무엇이 감지되었는지 확인할 수 있습니다.")
    else:
        print("여전히 객체가 감지되지 않았습니다. 이미지를 확인해주세요.")
```

```python
import cv2
import numpy as np
import os
from pathlib import Path

# ScoreCropUtils의 모든 함수들을 그대로 사용
def threshold_img(image):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    ret, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
    return image

def remove_noise(image):
    image = threshold_img(image)
    mask = np.zeros(image.shape, np.uint8)
    cnt, labels, stats, centroids = cv2.connectedComponentsWithStats(image)
    for i in range(1, cnt):
        x, y, w, h, area = stats[i]
        if w > image.shape[1] * 0.5:
            cv2.rectangle(mask, (x, y, w, h), (255, 0, 0), -1)
    masked_image = cv2.bitwise_and(image, mask)
    return masked_image

def remove_staves(image):
    height, width = image.shape
    staves = []
    for row in range(height):
        pixels = 0
        for col in range(width):
            pixels += (image[row][col] == 255)
        if pixels >= width * 0.5:
            if len(staves) == 0 or abs(staves[-1][0] + staves[-1][1] - row) > 1:
                staves.append([row, 0])
            else:
                staves[-1][1] += 1
    
    for staff in range(len(staves)):
        top_pixel = staves[staff][0]
        bot_pixel = staves[staff][0] + staves[staff][1]
        for col in range(width):
            if image[top_pixel - 1][col] == 0 and image[bot_pixel + 1][col] == 0:
                for row in range(top_pixel, bot_pixel + 1):
                    image[row][col] = 0
    return image, [x[0] for x in staves]

def normalization(image, staves, standard):
    if len(staves) < 6:
        print("⚠️ 감지된 오선 수가 부족합니다.")
        return image, staves
    
    avg_distance = 0
    lines = int(len(staves) / 5)
    for line in range(lines):
        for staff in range(4):
            staff_above = staves[line * 5 + staff]
            staff_below = staves[line * 5 + staff + 1]
            avg_distance += abs(staff_above - staff_below)
    
    denominator = len(staves) - lines
    if denominator == 0:
        print("⚠️ 나눗셈 0 오류 방지를 위해 avg_distance 계산을 건너뜁니다.")
        return image, [0 for _ in staves]
    
    avg_distance /= denominator
    height, width = image.shape
    weight = standard / avg_distance
    new_width = int(width * weight)
    new_height = int(height * weight)
    
    image = cv2.resize(image, (new_width, new_height))
    ret, image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    staves = [x * weight for x in staves]
    return image, staves

def weighted(value):
    standard = 10
    return int(value * (standard / 10))

def closing(image):
    kernel = np.ones((weighted(5), weighted(5)), np.uint8)
    image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    return image

def get_center(y, h):
    return (y + y + h) / 2

def object_detection(image, staves):
    lines = int(len(staves) / 5)
    objects = []
    closing_image = closing(image)
    cnt, labels, stats, centroids = cv2.connectedComponentsWithStats(closing_image)
    for i in range(1, cnt):
        (x, y, w, h, area) = stats[i]
        if w >= weighted(5) and h >= weighted(5):
            center = get_center(y, h)
            for line in range(lines):
                area_top = staves[line * 5] - weighted(20)
                area_bot = staves[(line + 1) * 5 - 1] + weighted(20)
                if area_top <= center <= area_bot:
                    objects.append([line, (x, y, w, h, area)])
    objects.sort()
    return image, objects

# 나머지 함수들도 모두 포함...
VERTICAL = True
HORIZONTAL = False

def get_line(image, axis, axis_value, start, end, length):
    if axis:
        points = [(i, axis_value) for i in range(start, end)]
    else:
        points = [(axis_value, i) for i in range(start, end)]
    pixels = 0
    for i in range(len(points)):
        (y, x) = points[i]
        pixels += (image[y][x] == 255)
        next_point = image[y + 1][x] if axis else image[y][x + 1]
        if next_point == 0 or i == len(points) - 1:
            if pixels >= weighted(length):
                break
            else:
                pixels = 0
    return y if axis else x, pixels

def stem_detection(image, stats, length):
    (x, y, w, h, area) = stats
    stems = []
    for col in range(x, x + w):
        end, pixels = get_line(image, VERTICAL, col, y, y + h, length)
        if pixels:
            if len(stems) == 0 or abs(stems[-1][0] + stems[-1][2] - col) >= 1:
                (x, y, w, h) = col, end - pixels + 1, 1, pixels
                stems.append([x, y, w, h])
            else:
                stems[-1][2] += 1
    return stems

def object_analysis(image, objects):
    for obj in objects:
        stats = obj[1]
        stems = stem_detection(image, stats, 30)
        direction = None
        if len(stems) > 0:
            if stems[0][0] - stats[0] >= weighted(5):
                direction = True
            else:
                direction = False
        obj.append(stems)
        obj.append(direction)
    return image, objects

def draw_staves(image, staves, color=(0, 0, 0), thickness=1):
    if len(image.shape) == 2:
        image = cv2.bitwise_not(image)
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    else:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.bitwise_not(gray)
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    
    for line_y in staves:
        cv2.line(image, (0, int(line_y)), (image.shape[1], int(line_y)), color, thickness)
    return image

def resize_with_padding(image, target_size=299, max_scale=4.0):
    h, w = image.shape[:2]
    scale = min(target_size / max(h, w), max_scale)
    new_w, new_h = int(w * scale), int(h * scale)
    resized = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    
    top = (target_size - new_h) // 2
    bottom = target_size - new_h - top
    left = (target_size - new_w) // 2
    right = target_size - new_w - left
    
    padded = cv2.copyMakeBorder(
        resized, top, bottom, left, right,
        borderType=cv2.BORDER_CONSTANT,
        value=[255, 255, 255]
    )
    return padded

class SmartMusicProcessor:
    def __init__(self, image_path, output_dir='smart_notes_output'):
        """
        친구의 ScoreCropUtils 기반 스마트 음표 추출기
        """
        self.image_path = image_path
        self.output_dir = Path(output_dir)
        self.output_dir.mkdir(exist_ok=True)
        
        self.original_image = None
        self.note_images = []
        self.saved_files = []
        
        print(f"출력 디렉토리: {self.output_dir.absolute()}")
    
    def load_image(self):
        """이미지 로드"""
        try:
            self.original_image = cv2.imread(self.image_path)
            if self.original_image is None:
                with open(self.image_path, 'rb') as f:
                    image_data = f.read()
                nparr = np.frombuffer(image_data, np.uint8)
                self.original_image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
            return True
        except Exception as e:
            print(f"이미지 로드 실패: {e}")
            return False
    
    def smart_object_filtering(self, objects):
        """
        스마트 객체 필터링 - 음표만 선별
        """
        print(f"필터링 전 객체 수: {len(objects)}")
        
        filtered_objects = []
        
        for obj in objects:
            line, stats, stems, direction = obj[0], obj[1], obj[2], obj[3]
            x, y, w, h, area = stats
            
            # 1. 크기 기반 필터링 (높은음자리표, 박자표 등 제외)
            if area > 3000:  # 너무 큰 기호들 제외
                print(f"큰 기호 제외: 면적={area}")
                continue
            
            if w > 80 or h > 80:  # 너무 큰 폭/높이 제외
                print(f"큰 객체 제외: 크기=({w}x{h})")
                continue
            
            # 2. 너무 작은 객체 제외
            if area < 50 or w < 8 or h < 8:
                continue
            
            # 3. 음표 특성 확인
            # - 기둥(stem)이 있거나
            # - 적절한 크기의 둥근 모양이거나
            # - 쉼표 모양인 경우
            has_stem = len(stems) > 0
            reasonable_size = 50 <= area <= 1500
            good_aspect_ratio = 0.2 <= (w/h) <= 5.0
            
            if has_stem or (reasonable_size and good_aspect_ratio):
                filtered_objects.append(obj)
                print(f"음표 후보 추가: 면적={area}, 크기=({w}x{h}), 기둥={len(stems)}개")
        
        print(f"필터링된 음표 수: {len(filtered_objects)}")
        return filtered_objects
    
    def extract_notes_only(self, image_with_staves, filtered_objects):
        """
        필터링된 음표들만 299x299로 추출하여 저장
        """
        # 이진화 이미지 생성 (크롭에 사용)
        binary = cv2.threshold(
            cv2.cvtColor(image_with_staves, cv2.COLOR_BGR2GRAY), 
            0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
        )[1]
        
        note_images = []
        saved_files = []
        
        for i, obj in enumerate(filtered_objects):
            try:
                line, stats = obj[0], obj[1]
                x, y, w, h, area = stats
                
                # 여백 추가하여 크롭
                margin = 10
                x1 = max(0, x - margin)
                y1 = max(0, y - margin)
                x2 = min(binary.shape[1], x + w + margin)
                y2 = min(binary.shape[0], y + h + margin)
                
                # 크롭
                cropped = binary[y1:y2, x1:x2]
                if cropped.size == 0:
                    continue
                
                # 299x299로 리사이즈 (패딩 포함)
                resized = resize_with_padding(cropped, target_size=299)
                
                # 파일 저장
                filename = f"smart_note_{i+1:03d}.png"
                filepath = self.output_dir / filename
                
                success = cv2.imwrite(str(filepath), resized)
                if success:
                    print(f"저장됨: {filename}")
                    saved_files.append(str(filepath))
                    note_images.append(resized)
                
            except Exception as e:
                print(f"객체 {i} 처리 중 오류: {e}")
                continue
        
        self.note_images = note_images
        self.saved_files = saved_files
        return note_images, saved_files
    
    def process_smart_score(self):
        """
        친구 코드 기반 스마트 악보 처리
        """
        print("=== 스마트 악보 처리 시작 ===")
        
        # 1. 이미지 로드
        if not self.load_image():
            return None
        print("✓ 이미지 로드 완료")
        
        # 2. 친구의 파이프라인 실행
        print("친구의 ScoreCropUtils 파이프라인 실행 중...")
        
        # 노이즈 제거
        image_1 = remove_noise(self.original_image)
        print("✓ 노이즈 제거 완료")
        
        # 오선 제거
        image_2, staves = remove_staves(image_1)
        print("✓ 오선 제거 완료")
        
        # 정규화
        image_3, staves = normalization(image_2, staves, standard=10)
        print("✓ 정규화 완료")
        
        # 객체 검출
        image_4, objects = object_detection(image_3, staves)
        print(f"✓ 객체 검출 완료: {len(objects)}개")
        
        # 객체 분석
        image_5, objects = object_analysis(image_4, objects)
        print("✓ 객체 분석 완료")
        
        # 3. 스마트 필터링 (새로운 단계)
        filtered_objects = self.smart_object_filtering(objects)
        print("✓ 스마트 필터링 완료")
        
        # 4. 오선 다시 그리기
        image_with_staves = draw_staves(image_5.copy(), staves)
        print("✓ 오선 복원 완료")
        
        # 5. 음표만 추출하여 저장
        note_images, saved_files = self.extract_notes_only(image_with_staves, filtered_objects)
        print("✓ 음표 추출 및 저장 완료")
        
        print("=== 처리 완료 ===")
        print(f"전체 감지된 객체 수: {len(objects)}")
        print(f"필터링된 음표 수: {len(filtered_objects)}")
        print(f"저장된 음표 이미지 수: {len(saved_files)}")
        print(f"출력 디렉토리: {self.output_dir.absolute()}")
        
        # numpy 배열로 저장
        if note_images:
            note_array = np.array(note_images)
            normalized_array = note_array.astype(np.float32) / 255.0
            model_input = np.expand_dims(normalized_array, axis=-1)
            
            numpy_path = self.output_dir / "smart_notes_array.npy"
            np.save(str(numpy_path), model_input)
            print(f"numpy 배열 저장: {numpy_path}")
            print(f"모델 입력 형태: {model_input.shape}")
        
        return {
            'note_images': note_images,
            'saved_files': saved_files,
            'filtered_objects': filtered_objects,
            'all_objects': objects,
            'staves': staves
        }

# 사용 예시
if __name__ == "__main__":
    # 스마트 처리기 생성
    image_path = r'C:\Users\kimsi\Desktop\명지대\딥러닝 과목 프로젝트\korea.jpg'
    processor = SmartMusicProcessor(image_path, 'smart_notes_output')
    
    # 스마트 파이프라인 실행
    results = processor.process_smart_score()
    
    if results and results['note_images']:
        print(f"\n=== 성공! ===")
        print(f"추출된 순수 음표 수: {len(results['note_images'])}")
        print("높은음자리표, 박자표 등은 스마트 필터링으로 제거되었습니다.")
        print("299x299 크기의 음표 이미지들이 저장되었습니다.")
    else:
        print("음표 추출에 실패했습니다.")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import cv2
import numpy as np
import math
from pathlib import Path
import matplotlib.pyplot as plt

# 한글 폰트 설정
import matplotlib.font_manager as fm
import platform

def setup_korean_font():
    system = platform.system()
    if system == 'Windows':
        plt.rcParams['font.family'] = 'Malgun Gothic'
    elif system == 'Darwin':
        plt.rcParams['font.family'] = 'AppleGothic'
    else:
        plt.rcParams['font.family'] = 'DejaVu Sans'
    plt.rcParams['axes.unicode_minus'] = False

setup_korean_font()

class PositionalEncoding2D(nn.Module):
    """
    SMT에서 사용하는 2D 위치 인코딩
    """
    def __init__(self, d_model, height, width):
        super().__init__()
        self.d_model = d_model
        
        # 2D 위치 인코딩 생성
        pe = torch.zeros(d_model, height, width)
        d_model_half = d_model // 2
        
        # Height 차원 인코딩
        div_term_h = torch.exp(torch.arange(0., d_model_half, 2) * 
                              -(math.log(10000.0) / d_model_half))
        pos_h = torch.arange(0., height).unsqueeze(1)
        pe[0:d_model_half:2, :, :] = torch.sin(pos_h * div_term_h).unsqueeze(2).repeat(1, 1, width)
        pe[1:d_model_half:2, :, :] = torch.cos(pos_h * div_term_h).unsqueeze(2).repeat(1, 1, width)
        
        # Width 차원 인코딩
        div_term_w = torch.exp(torch.arange(0., d_model_half, 2) * 
                              -(math.log(10000.0) / d_model_half))
        pos_w = torch.arange(0., width).unsqueeze(1)
        pe[d_model_half::2, :, :] = torch.sin(pos_w * div_term_w).unsqueeze(1).repeat(1, height, 1)
        pe[d_model_half+1::2, :, :] = torch.cos(pos_w * div_term_w).unsqueeze(1).repeat(1, height, 1)
        
        self.register_buffer('pe', pe.unsqueeze(0))
    
    def forward(self, x):
        # x: (batch_size, channels, height, width)
        return x + self.pe[:, :x.size(1), :x.size(2), :x.size(3)]

class CNNEncoder(nn.Module):
    """
    SMT의 CNN 기반 인코더
    """
    def __init__(self, input_channels=1, hidden_dim=256):
        super().__init__()
        
        # CNN 백본 (논문의 구현을 따름)
        self.conv_layers = nn.Sequential(
            # 첫 번째 블록
            nn.Conv2d(input_channels, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            
            # 두 번째 블록
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            
            # 세 번째 블록
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            
            # 네 번째 블록
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
        )
        
        # 출력 차원을 hidden_dim으로 맞춤
        self.projection = nn.Conv2d(256, hidden_dim, kernel_size=1)
        
        # 2D 위치 인코딩 (이미지 크기에 따라 조정 필요)
        self.pos_encoding = None
        
    def forward(self, x):
        # x: (batch_size, channels, height, width)
        features = self.conv_layers(x)
        features = self.projection(features)
        
        # 2D 위치 인코딩 적용
        if self.pos_encoding is None or self.pos_encoding.pe.shape[-2:] != features.shape[-2:]:
            self.pos_encoding = PositionalEncoding2D(
                features.size(1), features.size(2), features.size(3)
            ).to(features.device)
        
        features = self.pos_encoding(features)
        
        # (batch_size, channels, height, width) -> (batch_size, height*width, channels)
        batch_size, channels, height, width = features.shape
        features = features.view(batch_size, channels, -1).transpose(1, 2)
        
        return features

class TransformerDecoder(nn.Module):
    """
    SMT의 트랜스포머 디코더
    """
    def __init__(self, vocab_size, hidden_dim=256, num_layers=8, num_heads=4, max_length=1000):
        super().__init__()
        
        self.hidden_dim = hidden_dim
        self.vocab_size = vocab_size
        self.max_length = max_length
        
        # 토큰 임베딩
        self.token_embedding = nn.Embedding(vocab_size, hidden_dim)
        
        # 위치 인코딩
        self.pos_encoding = nn.Parameter(torch.randn(max_length, hidden_dim))
        
        # 트랜스포머 디코더 레이어
        decoder_layer = nn.TransformerDecoderLayer(
            d_model=hidden_dim,
            nhead=num_heads,
            dim_feedforward=hidden_dim * 4,
            dropout=0.1,
            batch_first=True
        )
        self.transformer_decoder = nn.TransformerDecoder(decoder_layer, num_layers)
        
        # 출력 프로젝션
        self.output_projection = nn.Linear(hidden_dim, vocab_size)
        
    def forward(self, encoder_output, target_sequence=None, max_length=None):
        if target_sequence is not None:
            # 훈련 모드
            seq_len = target_sequence.size(1)
            
            # 토큰 임베딩 + 위치 인코딩
            token_emb = self.token_embedding(target_sequence)
            pos_emb = self.pos_encoding[:seq_len].unsqueeze(0)
            decoder_input = token_emb + pos_emb
            
            # 마스크 생성 (자기회귀적 생성을 위해)
            tgt_mask = self.generate_square_subsequent_mask(seq_len).to(target_sequence.device)
            
            # 디코더 실행
            decoder_output = self.transformer_decoder(
                decoder_input, encoder_output, tgt_mask=tgt_mask
            )
            
            # 출력 프로젝션
            logits = self.output_projection(decoder_output)
            return logits
        else:
            # 추론 모드
            return self.generate(encoder_output, max_length or 100)
    
    def generate_square_subsequent_mask(self, sz):
        """자기회귀적 마스크 생성"""
        mask = torch.triu(torch.ones(sz, sz), diagonal=1)
        mask = mask.masked_fill(mask == 1, float('-inf'))
        return mask
    
    def generate(self, encoder_output, max_length):
        """자기회귀적 생성"""
        batch_size = encoder_output.size(0)
        device = encoder_output.device
        
        # 시작 토큰 (여기서는 1로 가정)
        generated = torch.ones(batch_size, 1, dtype=torch.long, device=device)
        
        for _ in range(max_length - 1):
            # 현재까지 생성된 시퀀스로 다음 토큰 예측
            logits = self.forward(encoder_output, generated)
            next_token = torch.argmax(logits[:, -1:], dim=-1)
            generated = torch.cat([generated, next_token], dim=1)
            
            # 종료 토큰 확인 (여기서는 2로 가정)
            if (next_token == 2).all():
                break
        
        return generated

class SheetMusicTransformer(nn.Module):
    """
    Sheet Music Transformer 메인 모델
    """
    def __init__(self, vocab_size=100, hidden_dim=256, num_layers=8, num_heads=4):
        super().__init__()
        
        self.encoder = CNNEncoder(input_channels=1, hidden_dim=hidden_dim)
        self.decoder = TransformerDecoder(
            vocab_size=vocab_size,
            hidden_dim=hidden_dim,
            num_layers=num_layers,
            num_heads=num_heads
        )
        
    def forward(self, images, target_sequences=None):
        # 인코더로 이미지 특징 추출
        encoder_output = self.encoder(images)
        
        # 디코더로 시퀀스 생성
        if target_sequences is not None:
            # 훈련 모드
            return self.decoder(encoder_output, target_sequences)
        else:
            # 추론 모드
            return self.decoder(encoder_output)

class MusicNotationVocabulary:
    """
    음악 기호 어휘 사전
    """
    def __init__(self):
        # 기본 토큰들
        self.special_tokens = {
            '<PAD>': 0,
            '<SOS>': 1,  # Start of sequence
            '<EOS>': 2,  # End of sequence
            '<UNK>': 3,  # Unknown
        }
        
        # 음표 기호들 (88건반 기준)
        self.note_tokens = {}
        for i in range(88):
            self.note_tokens[f'NOTE_{i}'] = len(self.special_tokens) + i
        
        # 쉼표 기호들
        self.rest_tokens = {
            'WHOLE_REST': len(self.special_tokens) + 88,
            'HALF_REST': len(self.special_tokens) + 89,
            'QUARTER_REST': len(self.special_tokens) + 90,
            'EIGHTH_REST': len(self.special_tokens) + 91,
            'SIXTEENTH_REST': len(self.special_tokens) + 92,
        }
        
        # 전체 어휘 사전 구성
        self.vocab = {}
        self.vocab.update(self.special_tokens)
        self.vocab.update(self.note_tokens)
        self.vocab.update(self.rest_tokens)
        
        # 역방향 매핑
        self.idx_to_token = {v: k for k, v in self.vocab.items()}
        
        self.vocab_size = len(self.vocab)
    
    def encode(self, tokens):
        """토큰 리스트를 인덱스 리스트로 변환"""
        return [self.vocab.get(token, self.vocab['<UNK>']) for token in tokens]
    
    def decode(self, indices):
        """인덱스 리스트를 토큰 리스트로 변환"""
        return [self.idx_to_token.get(idx, '<UNK>') for idx in indices]

class KoreanScoreDataset(Dataset):
    """
    한국 악보 데이터셋
    """
    def __init__(self, image_paths, annotations=None, transform=None):
        self.image_paths = image_paths
        self.annotations = annotations
        self.transform = transform
        self.vocab = MusicNotationVocabulary()
        
    def __len__(self):
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        # 이미지 로드
        image_path = self.image_paths[idx]
        image = cv2.imread(str(image_path), cv2.IMREAD_GRAYSCALE)
        
        if image is None:
            # 한글 경로 처리
            with open(image_path, 'rb') as f:
                image_data = f.read()
            nparr = np.frombuffer(image_data, np.uint8)
            image = cv2.imdecode(nparr, cv2.IMREAD_GRAYSCALE)
        
        # 이미지 전처리
        if self.transform:
            image = self.transform(image)
        else:
            # 기본 전처리
            image = cv2.resize(image, (512, 256))  # 고정 크기로 리사이즈
            image = image.astype(np.float32) / 255.0
            image = torch.from_numpy(image).unsqueeze(0)  # 채널 차원 추가
        
        # 어노테이션이 있는 경우
        if self.annotations and idx < len(self.annotations):
            annotation = self.annotations[idx]
            # 어노테이션을 토큰 시퀀스로 변환
            tokens = ['<SOS>'] + annotation + ['<EOS>']
            token_indices = torch.tensor(self.vocab.encode(tokens), dtype=torch.long)
            return image, token_indices
        
        return image

class SMTTrainer:
    """
    SMT 모델 훈련기
    """
    def __init__(self, model, vocab, device='cuda' if torch.cuda.is_available() else 'cpu'):
        self.model = model.to(device)
        self.vocab = vocab
        self.device = device
        self.criterion = nn.CrossEntropyLoss(ignore_index=vocab.vocab['<PAD>'])
        self.optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
        
    def train_step(self, images, target_sequences):
        self.model.train()
        
        images = images.to(self.device)
        target_sequences = target_sequences.to(self.device)
        
        # 입력과 타겟 분리 (teacher forcing)
        input_seq = target_sequences[:, :-1]
        target_seq = target_sequences[:, 1:]
        
        # 순전파
        logits = self.model(images, input_seq)
        
        # 손실 계산
        loss = self.criterion(logits.reshape(-1, logits.size(-1)), target_seq.reshape(-1))
        
        # 역전파
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        
        return loss.item()
    
    def predict(self, image):
        """단일 이미지에 대한 예측"""
        self.model.eval()
        
        with torch.no_grad():
            if isinstance(image, np.ndarray):
                image = torch.from_numpy(image).unsqueeze(0).unsqueeze(0).float()
            
            image = image.to(self.device)
            
            # 예측 수행
            predicted_indices = self.model(image)
            
            # 토큰으로 변환
            predicted_tokens = self.vocab.decode(predicted_indices[0].cpu().numpy())
            
            return predicted_tokens

def create_demo_model():
    """
    데모용 SMT 모델 생성
    """
    vocab = MusicNotationVocabulary()
    model = SheetMusicTransformer(
        vocab_size=vocab.vocab_size,
        hidden_dim=256,
        num_layers=6,  # 데모용으로 축소
        num_heads=4
    )
    
    trainer = SMTTrainer(model, vocab)
    
    return model, trainer, vocab

def test_smt_on_korean_score(image_path):
    """
    한국 악보에 SMT 모델 테스트
    """
    print("=== Sheet Music Transformer 테스트 ===")
    
    # 모델 생성
    model, trainer, vocab = create_demo_model()
    
    # 이미지 로드 및 전처리
    try:
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if image is None:
            with open(image_path, 'rb') as f:
                image_data = f.read()
            nparr = np.frombuffer(image_data, np.uint8)
            image = cv2.imdecode(nparr, cv2.IMREAD_GRAYSCALE)
        
        print(f"✓ 이미지 로드 완료: {image.shape}")
        
        # 전처리
        processed_image = cv2.resize(image, (512, 256))
        processed_image = processed_image.astype(np.float32) / 255.0
        
        print("✓ 이미지 전처리 완료")
        
        # 예측 수행 (랜덤 초기화된 모델이므로 의미있는 결과는 아님)
        predicted_tokens = trainer.predict(processed_image)
        
        print("✓ SMT 예측 완료")
        print(f"예측된 토큰 수: {len(predicted_tokens)}")
        print(f"예측 결과 (처음 10개): {predicted_tokens[:10]}")
        
        # 시각화
        plt.figure(figsize=(15, 8))
        
        plt.subplot(1, 2, 1)
        plt.imshow(image, cmap='gray')
        plt.title('원본 악보 이미지')
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        plt.imshow(processed_image, cmap='gray')
        plt.title('전처리된 이미지 (512x256)')
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        # 모델 구조 정보
        total_params = sum(p.numel() for p in model.parameters())
        trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
        
        print(f"\n=== 모델 정보 ===")
        print(f"총 파라미터 수: {total_params:,}")
        print(f"훈련 가능한 파라미터 수: {trainable_params:,}")
        print(f"어휘 사전 크기: {vocab.vocab_size}")
        print(f"인코더: CNN 기반")
        print(f"디코더: Transformer (6 layers, 4 heads)")
        
        return model, trainer, vocab, predicted_tokens
        
    except Exception as e:
        print(f"오류 발생: {e}")
        return None

# 실행 예시
if __name__ == "__main__":
    # 한국 악보 이미지 경로
    image_path = r'C:\Users\kimsi\Desktop\명지대\딥러닝 과목 프로젝트\korea.jpg'
    
    # SMT 모델 테스트
    results = test_smt_on_korean_score(image_path)
    
    if results:
        model, trainer, vocab, predictions = results
        print("\n=== SMT 구현 완료 ===")
        print("주의: 현재는 랜덤 초기화된 모델입니다.")
        print("실제 사용을 위해서는 대량의 악보 데이터로 훈련이 필요합니다.")
        print("\n다음 단계:")
        print("1. 악보-텍스트 쌍 데이터셋 구축")
        print("2. 모델 훈련 (수주~수개월)")
        print("3. 하이퍼파라미터 튜닝")
        print("4. 성능 평가 및 개선")
```
